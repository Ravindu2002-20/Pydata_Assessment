```python
import dash
from dash import html, dcc, Output, Input, State
import base64
from pathlib import Path 

# Initialize the app
app = dash.Dash(__name__)
app.title = "Wine Review Dashboard"

image_path = Path("Images/ai-generated-fruits-grapes-wine-glasses-and-wine-bottles-on-a-classic-vintage-wooden-background-free-photo.jpeg")
encoded_image = base64.b64encode(open(image_path, "rb").read()).decode()

background_image_url = f"data:image/jpeg;base64,{encoded_image}"

# App Layout
app.layout = html.Div(
    style={
        'backgroundImage': f'url("{background_image_url}")',
        'backgroundSize': 'cover',
        'backgroundPosition': 'center',
        'backgroundRepeat': 'no-repeat',
        'backgroundAttachment': 'fixed',
        'minHeight': '100vh',
        'color': 'white',
    },
    children=[
        # Top Navbar
        html.Div(
            style={
                'height': '60px',
                'background': 'linear-gradient(to bottom, #3f0400, #9e3524, #360700)',
                'display': 'flex',
                'alignItems': 'center',
                'padding': '0 20px'
            },
            children=[
                html.Button("☰ Menu", id="menu-button", n_clicks=0, style={
                    'fontSize': '20px',
                    'background': 'none',
                    'border': 'none',
                    'color': 'white',
                    'cursor': 'pointer'
                }),
                html.H1("Wine Review Dashboard", style={
                    'margin': '0 auto',
                    'fontSize': '24px'
                })
            ]
        ),

        # Full container
        html.Div(
            style={
                'display': 'flex',
                'height': 'calc(100vh - 60px)'
            },
            children=[
                html.Div(
                    id="sidebar",
                    style={
                        'width': '250px',
                        'background': 'url("/Images/Sidebar Image.jpeg")',
                        'backgroundSize': 'cover',
                        'backgroundPosition': 'center',
                        'backgroundRepeat': 'no-repeat',
                        'opacity': 1,
                        'padding': '20px',
                        'boxShadow': '2px 0 5px rgba(0,0,0,0.1)',
                        'display': 'block'
                    },
                    children=[
                        html.H2("Menu", style={'textAlign': 'center', 'color': '#fff'}),
                        html.Hr(),

                        html.Ul([
                            html.Li(html.Button("Home", id="home-btn", n_clicks=0, className="sidebar-btn", style={
                                'margin': '10px 0',
                                'padding': '10px',
                                'background': 'rgba(255,255,255,0.1)',
                                'border': '1px solid #fff',
                                'borderRadius': '5px',
                                'textAlign': 'center',
                                'width': '100%',
                                'color': '#fff',
                                'cursor': 'pointer',
                                'transition': '0.2s'
                            })),
                            html.Li(html.Button("Analytics", id="analytics-btn", n_clicks=0, className="sidebar-btn", style={
                                'margin': '10px 0',
                                'padding': '10px',
                                'background': 'rgba(255,255,255,0.1)',
                                'border': '1px solid #fff',
                                'borderRadius': '5px',
                                'textAlign': 'center',
                                'width': '100%',
                                'color': '#fff',
                                'cursor': 'pointer',
                                'transition': '0.2s'
                            })),
                            html.Li(html.Button("Reviews", id="reviews-btn", n_clicks=0, className="sidebar-btn", style={
                                'margin': '10px 0',
                                'padding': '10px',
                                'background': 'rgba(255,255,255,0.1)',
                                'border': '1px solid #fff',
                                'borderRadius': '5px',
                                'textAlign': 'center',
                                'width': '100%',
                                'color': '#fff',
                                'cursor': 'pointer',
                                'transition': '0.2s'
                            })),
                        ], style={'listStyleType': 'none', 'padding': 0})
                    ]
                ),

                # Main Content
                html.Div(
                    id="main-content",
                    style={
                        'flex': 1,
                        'padding': '20px',
                        'backgroundColor': 'rgba(0,0,0,0.5)',
                        'borderRadius': '10px',
                        'margin': '20px'
                    },
                    children=[
                        html.H2("Welcome to Wine Review"),
                        html.P("Select a menu item to see more!")
                    ]
                )
            ]
        )
    ]
)


# Callbacks
@app.callback(
    Output('sidebar', 'style'),
    Input('menu-button', 'n_clicks'),
    State('sidebar', 'style')
)
def toggle_sidebar(n_clicks, sidebar_style):
    if n_clicks % 2 == 1:
        sidebar_style['display'] = 'none'
    else:
        sidebar_style['display'] = 'block'
    return sidebar_style

@app.callback(
    Output('main-content', 'children'),
    [Input('home-btn', 'n_clicks'),
     Input('analytics-btn', 'n_clicks'),
     Input('reviews-btn', 'n_clicks'),
    ]
)
def display_page(home, analytics, reviews):
    ctx = dash.callback_context

    if not ctx.triggered:
        return [html.H2("Welcome to Wine Review"), html.P("Select a menu item to see more!")]
    else:
        button_id = ctx.triggered[0]['prop_id'].split('.')[0]

    if button_id == "home-btn":
        return get_home_page()
    elif button_id == "analytics-btn":
        return [html.H2("Analytics"), html.P("Data insights and graphs about wine.")]
    elif button_id == "reviews-btn":
        return [html.H2("Reviews"), html.P("Individual wine review details.")]
    else:
        return [html.H2("Welcome to Wine Review"), html.P("Select a menu item to see more!")]

# Run the app
if __name__ == "__main__":
    port = find_free_port()
    print(f"Running on port {port}...")
    app.run(debug=False, port=port)
```

```python
# Home page
from datetime import datetime

def get_home_page():
    now = datetime.now().strftime("%A, %B %d, %Y — %I:%M %p")
    return [
        html.H1("Welcome to the Wine Review Dashboard!", style={"fontSize": "38px"}),
        html.P("Welcome to the Wine Review Classifier Dashboard, a data-driven exploration of 500+ wine reviews powered by Hugging Face’s zero-shot AI model. This interactive tool categorizes tasting notes into key themes such as food pairings, taste profiles, value assessments, and other insights, helping sommeliers, enthusiasts, and data lovers uncover patterns in wine feedback. Built with Python and modern visualization libraries, the dashboard transforms raw reviews into an engaging, analytical experience complete with flavor word clouds, vintage trends, and real-time classification. Whether you're refining your palate or studying consumer preferences, this project blends oenology with AI to reveal the stories behind every sip. Cheers to data-inspired discoveries.",style={'marginBottom': '20px'}),
        html.P(f"Date & Time: {now}", style={"fontStyle": "italic"})
    ]
```

```python
import socket
from dash import Dash

def find_free_port():
    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
        s.bind(('', 0))  # 0 means to find any free port
        return s.getsockname()[1]

app = Dash(__name__)
app.title = "Wine Review Dashboard"
```
